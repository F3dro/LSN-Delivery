## <span style="color:black"> Esercizio 05.1 </span>

L'idea dell'esercizio è quella di sfruttare l'algoritmo di Metropolis, che permette di campionare una generica densità di probabilità, per campionare la densità di probabilità dei primi stati dell'atomo di idrogeno. Infatti data una funzione d'onda $\Psi_{n,l,m}(r,\theta, \phi)$ si ha che la densità di probabilità associata è data da $\left|\Psi_{n,l,m}(r,\theta, \phi)\right|^2$. In particolare per l'atomo di igrogeno  due stati i 1s e 2p sono caratterizzati dalle funzioni d'onda:
$$\Psi_{100}(r,\theta,\phi)=\textit{N}e^{-r/a_0} \quad \mathrm{e} \quad \Psi_{210}(r,\theta,\phi)=\textit{N'}re^{-r/2a_0}cos(\theta)$$ 
Si è evitato di calcolarne la normalizzazione in quanto nell'algoritmo di Metropolis la dipendenza dalla densità di probabilità compare esclusivamente in un rapporto tra la stessa densità valutata in due diversi punti, motivo per il quale la normalizzazione si semplifica e non rappresenta quindi un elemento necessario da conoscere per campionare la denistà di probabilità desiderata.

Lo scopo di questo esercizio è dunque quello di campionare le due densità di probabilità $\left|\Psi_{100}(r,\theta, \phi)\right|^2$ e $\left|\Psi_{210}(r,\theta, \phi)\right|^2$ in modo poi da stimare i valori medi delle distanze dal centro $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ e $\left\langle r \right\rangle_{\Psi_{2,1,0}}$.

L'algoritmo di Metropolis in particolare consiste nel proporre una mossa $x'$ generata secondo una distribuzione $T(x'|x_n)$ e accettarla con una probabilità $\alpha$ tale che, detto $r$ numero casuale generato uniformemente in [0,1], se $r\le\alpha \Rightarrow x_{n+1}=x'$ altrimenti $x_{n+1}=x_n$. In particolare la probabilità $\alpha$ è data da 
$$\alpha=A(x'|x_n)=min\left\{1,\frac{T(x_n|x')p(x')}{T(x'|x_n)p(x_n)} \right\}=min\left\{1,\frac{p(x')}{p(x_n)} \right\}$$
dove l'ultima uguaglianza vale nel caso che $T(x'|x_n)$ sia una distribuzione simmetrica, ovvero che valga $T(x|y)=T(y|x)$.

E' stato riportata nel grafico sottostante la visualizzazione delle posizioni campionate in questo modo per i due stati 1s e 2p dell'idrogeno, in particolare utilizzando $10^5$ steps e una distribuzione $T(x'|x_n)$ uniforme. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X, Y, Z=np.loadtxt('Results/output_GS.dat', delimiter=' ', usecols=(1,2,3), unpack='true')

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel(r'$x/a_0$')
ax.set_ylabel(r'$y/a_0$')
ax.set_zlabel(r'$z/a_0$')
ax.view_init(10, 30)
plt.title('Densità di probabilità dello stato 1s')
plt.show()

X, Y, Z=np.loadtxt('Results/output_2p.dat', delimiter=' ', usecols=(1,2,3), unpack='true')

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel(r'$x/a_0$')
ax.set_ylabel(r'$y/a_0$')
ax.set_zlabel(r'$z/a_0$')
ax.view_init(10, 30)
plt.title('Densità di probabilità dello stato 2p')
plt.show()
```

Un'altra questione importante nell'utilizzo dell'algoritmo di Metropolis è la scelta del passo $\delta$ con cui proporre i nuovi movimenti: qualora esso sita troppo piccolo l'accettazione risulterà molto elevata poichè $x'\simeq x_n \Rightarrow p(x')\simeq p(x_n)$ e di conseguenza si avrà un'elevata correlazione tra i dati. D'altro  canto scegliere un passo troppo grande causa un'accettazione, di norma molto piccola, per cui la convergenza sarà molto lenta e quindi il processo poco efficiente. Una buona regola empirica è quella di regolare il passo affinchè l'accettazione, se possibile, sia intorno al 50%.  
Per fare questo è quindi necessaria una fase di equilibrazione: in queste simulazioni è stata fatta a mano per regolare, caso per caso, il passo $\delta$ tale che $\alpha \simeq 50\%$.

E' necessario prestare attenzione anche al numero di blocchi, e quiundi di steps per blocco da utilizzare: per evitare una correlazione tra i dati di ogni blocco bisogna evitare di scegliere pochi steps per ogni blocco, ma allo stesso tempo occorre avere un numero di blocchi sufficienti per avere una buona stima del valore finale. Per questi motivi, nelle simulazioni seguenti, sono stati effettuati $10^6$ Monte Carlo steps complessivi e $200$ blocchi, da cui risultano $5000$ steps per blocco.

Dal momento che l'unica richiesta per ora effettuata sulla distribuzione $T(x'|x_n)$ è la proprietà di simmetria, è stata utilizzata anche una distribuzione normale. I risultati sono presentati nei grafici successivi e confermano che non si presenta nessun cambiamento significativo nell'utilizzo di una distribuzione rispetto all'altra.

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(15, 15))
fig.suptitle('Distance mean value for different states', y='0.92', size='20', weight='20')
gs = gridspec.GridSpec(2, 2, figure=fig)

x1, f1, error1 = np.loadtxt('uniform/output_rmean_GS.dat', usecols=(0,1,2), unpack='true')
x2, f2, error2 = np.loadtxt('gaussian/output_rmean_GS.dat', usecols=(0,1,2), unpack='true')
ax = fig.add_subplot(gs[0, 0])
ax.errorbar(x1, f1, yerr=error1, c='xkcd:blue', ecolor='xkcd:aqua', label='Points from uniform distribution')
ax.errorbar(x2, f2, yerr=error2, c='xkcd:orange', ecolor='xkcd:yellow', label='Points from gaussian distribution')
ax.plot(1.5 +x1-x1, c='xkcd:dark blue', label='Expected value')
ax.set_xlabel("Number of blocks")
ax.set_ylabel(r'$<r>_{\Psi_{1,0,0}}/a_0$')
plt.title("1s")
ax.legend()
ax.grid()

x1, f1, error1 = np.loadtxt('uniform/output_rmean_2p.dat', usecols=(0,1,2), unpack='true')
x2, f2, error2 = np.loadtxt('gaussian/output_rmean_2p.dat', usecols=(0,1,2), unpack='true')
ax2 = fig.add_subplot(gs[0, 1])
ax2.errorbar(x1, f1, yerr=error1, c='xkcd:blue', ecolor='xkcd:aqua', label='Points from uniform distribution')
ax2.errorbar(x2, f2, yerr=error2, c='xkcd:orange', ecolor='xkcd:yellow', label='Points from gaussian distribution')
ax2.plot(5 +x1-x1, c='xkcd:dark blue', label='Expected value')
ax2.set_xlabel("Number of blocks")
ax2.set_ylabel(r'$<r>_{\Psi_{2,1,0}}/a_0$')
plt.title("2p")
ax2.legend()
ax2.grid()

plt.show()
```

Nei grafici appena presentati, dopo una fase di equilibrazione, sono stati scelti valori ottimali di partenza. Si nota che già da circa 50 steps Monte Carlo si ha una ottima convergenza al risultato atteso teoricamente ($1.5 a_0$ per lo stato 1s e $5a_0$ per lo stato 2p).  
Ci si è chiesti cosa succeda se il punto di partenza non sia quello ottimale, per cui nei grafici successivi sono stati presentati i risultati ottenuti avendo utilizzato punti di partenza molto lontani da quelli di equilibrio, nominalmente $25a_0$ per lo stato 1s e $30a_0$ per lo stato 2p. Quello che sorprende è che la velocità di convergenza risulta pressochè inalterata 

```python
fig = plt.figure(figsize=(15, 15))
fig.suptitle('Distance mean value for different states starting far from equilibrium distance', y='0.92', size='20', weight='20')
gs = gridspec.GridSpec(2, 2, figure=fig)

x1, f1, error1 = np.loadtxt('far_start/output_rmean_GS.dat', usecols=(0,1,2), unpack='true')
ax = fig.add_subplot(gs[0, 0])
ax.errorbar(x1, f1, yerr=error1, c='xkcd:blue', ecolor='xkcd:aqua', label=r'tarted at $25a_0$')
ax.plot(1.5 +x1-x1, c='xkcd:orange', label='Expected value')
ax.set_xlabel("Number of blocks")
ax.set_ylabel(r'$<r>_{\Psi_{1,0,0}}/a_0$')
plt.title("1s")
ax.legend()
ax.grid()

x1, f1, error1 = np.loadtxt('far_start/output_rmean_2p.dat', usecols=(0,1,2), unpack='true')
ax2 = fig.add_subplot(gs[0, 1])
ax2.errorbar(x1, f1, yerr=error1, c='xkcd:blue', ecolor='xkcd:aqua', label=r'Started at $30a_0$')
ax2.plot(5 +x1-x1, c='xkcd:orange', label='Expected value')
ax2.set_xlabel("Number of blocks")
ax2.set_ylabel(r'$<r>_{\Psi_{2,1,0}}/a_0$')
plt.title("2p")
ax2.legend()
ax2.grid()

plt.show()
```
