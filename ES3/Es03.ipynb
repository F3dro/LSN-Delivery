##### Pianificazione del prezzo di opzioni europee
Per ottenere il valore attuale dell'opzione, bisoga innanzitutto stimare il profitto al tempo $T$ della scadenza dell'opzione, che e' data per le Call-Options da $$\bigl(S(T)-K\bigr)^+ := max\bigl[0, S(T)-K \bigr]$$
A questo punto il costo dell'opzione si ottiene scontando il guadagno "risk free" dato dal fattore $r$, per cui il prezzo dell'opzione e' dato da:
$$\left<e^{-rT}\left(S(T)-K\right)^+\right>$$

dove $S(t)$ puo' essere stimato ipotizzando un moto browniano geometrico in modo diretto al tempo $T$ oppure in modo progressivo dividento l'intervallo $[0, T]$ in $N$ parti e valutando $S(t_{i+1})$ a partire da $S(t_i)$ per $i=0,...N$.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x1, f1, error1 = np.loadtxt('Es03.1.1.res', usecols=(0,1,2), unpack='true')
x2, f2, error2 = np.loadtxt('Es03.1.2.res', usecols=(0,1,2), unpack='true')

plt.errorbar(x1, f1, yerr=error1, label='Calcolo diretto')
plt.errorbar(x2, f2, yerr=error2, label='Calcolo progressivo')
plt.plot(x1, 14.975790778311286-x1+x1, 'g', label='Valore analitico')
plt.xlabel("Number of blocks")
plt.ylabel("Option price")
plt.title("Call-option")
plt.legend()
plt.grid()
plt.show()
```

Dal grafico si pu√≤ effettivamente osservare l'esattezza del metodo, infatti entrambi i metodi di calcolo conducono statisticamente al medesimo risultato.

Per le Put-Options si puo' procedere in modo analogo, ricordando pero' che il profitto al tempo della scadenza e' dato da $\left(K-S(T)\right)^+$

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x1, f1, error1 = np.loadtxt('Es03.1.3.res', usecols=(0,1,2), unpack='true')
x2, f2, error2 = np.loadtxt('Es03.1.4.res', usecols=(0,1,2), unpack='true')

plt.errorbar(x1, f1, yerr=error1, label='Calcolo diretto')
plt.errorbar(x2, f2, yerr=error2, label='Calcolo progressivo')
plt.plot(x1, 5.4595325819072364-x1+x1, 'g', label='Valore analitico')
plt.xlabel("Number of blocks")
plt.ylabel("Option price")
plt.title("Put-option")
plt.legend()
plt.grid()
plt.show()
```

Anche in questo caso √® verificata la convergenza statistica del metodo al valore atteso dai conti analitici.
