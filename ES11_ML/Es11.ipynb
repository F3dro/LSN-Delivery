## <span style="color:black"> Esercizio 11.1: Fit lineare </span>

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import tensorflow as tf
from tensorflow import keras
from keras import optimizers

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Flatten
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
```

L'idea dell'esercizio è studiare come i parametri di allenamento di una rete neurale influenzano la bontà del fit da essa effettuato.  
In particolare in questo esercizio si è studiata l'influenza di fattori quali il numero di epoche di allenamento ($N_{epochs}$), il numero di punti su cui è stato effettuato l'allenamento ($N_{train}$), e il rumore con cui sono distribuiti questi punti rispetto alla funzione scelta. Tale funzione è la retta:
$$ f(x)=2x+1 $$
di conseguenza il fit corrisponde ad una regressione lineare. In questo caso specifico, il compito può essere svolto da un solo neurone, quindi questo permette di poter trascurare momentaneamente le caratteristiche della rete neurale per concentrarsi sui parametri dell'allenamento.  
E' stata scritta la funzione <font face = "Courier"> `do_it` </font> che crea una rete neurale con un unico neurone, la allena per <font face = "Courier"> in_epochs </font> epoche sulla base di <font face = "Courier"> n_train </font> punti distribuiti con rumore gaussiano di varianza <font face = "Courier"> in_sigma </font> attorno alla funzione $f(x)$ sopra definita, e inoltre la testa generando <font face = "Courier"> n_valid </font> punti secondo il modello così creato, in modo da poterne valutare la bontà.

```python
def do_it (n_train, n_valid, in_sigma, in_epochs):
    
    # target parameters of f(x) = m*x + b
    m = 2 # slope
    b = 1 # intersect
    # generate training inputs
    np.random.seed(0)
    x_train = np.random.uniform(-1, 1, n_train)
    x_valid = np.random.uniform(-1, 1, n_valid)
    x_valid.sort()
    y_target = m * x_valid + b # ideal (target) linear function
    
    sigma = in_sigma # noise standard deviation, for the moment it is absent
    y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
    y_valid = np.random.normal(m * x_valid + b, sigma)
    
    model = tf.keras.Sequential()
    model.add(Dense(1, input_shape=(1,)))
    
    # compile the model choosing optimizer, loss and metrics objects
    model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
    # fit the model using training dataset
    # over 10 epochs of 32 batch size each
    # report training progress against validation data
    history = model.fit(x=x_train, y=y_train, 
                        batch_size=32, epochs=in_epochs,
                        shuffle=True, # a good idea is to shuffle input before at each epoch
                        validation_data=(x_valid, y_valid),
                        verbose=0);
    # return weights
    print(model.get_weights())
    
    x_predicted = np.random.uniform(-1, 1, 100)
    y_predicted = model.predict(x_predicted)
    plt.scatter(x_predicted, y_predicted,color='r')
    plt.plot(x_valid, y_target)
    plt.grid(True)
    plt.show()
    return model
```

I risultati ottenuti variando i parametri di input della funzione sono riportati nella cella seguente. Si nota in generale un miglioramento del fit con il numero di punti su cui viene fatto l'allenamento. Il numero di epoche non è altrettanto significativo per un solo neurone. Si nota invece che per ottenere risultati ottimali è necessario che il rumore non sia nullo, in modo che la rete possa imparare in modo più generale.

```python
n_train_vec = [100, 250, 500, 1000]
n_valid_vec = [10, 25, 50, 100]
sigma_vec = [0, 0.2, 0.5, 1]
in_epochs_vec = [30, 50, 70, 100]

#for i in range(4):
#    for j in range(3):
        #fig.add_subplot(gs[i, j])
 #       plt.title('n_train='+str(n_train_vec[i])+', epochs='+str(in_epochs_vec[j]))
 #       do_it(n_train_vec[i], n_valid_vec[i], 0, in_epochs_vec[j])


plt.title('n_train=100, epochs=30, sigma=0')
do_it(n_train_vec[0], n_valid_vec[0], sigma_vec[0], in_epochs_vec[0])

plt.title('n_train=250, epochs=50, sigma=0')
do_it(n_train_vec[1], n_valid_vec[1], sigma_vec[0], in_epochs_vec[2])

plt.title('n_train=250, epochs=50, sigma=1')
do_it(n_train_vec[1], n_valid_vec[1], sigma_vec[3], in_epochs_vec[2])

plt.title('n_train=1000, epochs=30, sigma=0')
do_it(n_train_vec[3], n_valid_vec[3], sigma_vec[0], in_epochs_vec[0])

plt.title('n_train=1000, epochs=30, sigma=1')
do_it(n_train_vec[3], n_valid_vec[3], sigma_vec[3], in_epochs_vec[0])

plt.title('n_train=1000, epochs=30, sigma=10')
do_it(n_train_vec[3], n_valid_vec[3], 10, in_epochs_vec[0])
```

A questo punto si è provato ad estendere il modello per realizzare un fit polinomiale, della funzione:
$$ f(x)=4-3x-2x^2+3x^3 $$
nell'intervallo $x \in [-1,1]$. Analogamente al caso precedente è stata scritta una funzione <font face = "Courier"> `do_poly` </font>, la quale questa volta crea una rete neurale più complessa, ovvero composta da più neuroni su un layer o addirittura più layer composti da più neuroni. Sono state esplorate diverse configurazioni per la struttura della rete neurale salvando in alcuni modelli i risultati principali, esposti nelle celle successive. La funzione di attivazione che si è alla fine deciso di utilizzare è la tangente iperbolica, in quanto garantisce una convergenza migliore. E' inoltre stato osservato che per fit più complessi (ad esempio in più dimensioni come nell'esercizio successivo), il fatto che questa funzione assuma anche valori negativi è una caratteristica imprescindibile, senza la quale difficilmente è possibile garantire un buon apprendimento della rete.

```python
def do_poly (n_train, n_valid, in_sigma, in_epochs):
    a=4
    b=3
    c=2
    d=3
    # generate training inputs
    np.random.seed(0)
    x_train = np.random.uniform(-1, 1, n_train)
    x_valid = np.random.uniform(-1, 1, n_valid)
    x_valid.sort()
    y_target = a - b * x_valid - c * np.power(x_valid,2) + d * np.power(x_valid,3) # ideal (target) linear function
    
    sigma = in_sigma # noise standard deviation, for the moment it is absent
    y_train = np.random.normal( a - b * x_train - c * np.power(x_train,2) + d * np.power(x_train,3), sigma) # actual measures from which we want to guess regression parameters
    y_valid = np.random.normal( a - b * x_valid - c * np.power(x_valid,2) + d * np.power(x_valid,3), sigma)
    
    
    plt.plot(x_valid, y_target)
    plt.scatter(x_valid, y_valid, color='r')
    plt.grid(True); plt.show()


    model = tf.keras.Sequential()
    model.add(Dense(4, input_shape=(1,), activation='tanh'))
    model.add(Dense(6, activation='tanh'))
    model.add(Dense(7, activation='tanh'))
    model.add(Dense(5, activation='tanh'))
    model.add(Dense(4, activation='tanh'))
    model.add(Dense(2, activation='tanh'))
    model.add(Dense(1))
    
    # compile the model choosing optimizer, loss and metrics objects
    model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
    # fit the model using training dataset
    # over 10 epochs of 32 batch size each
    # report training progress against validation data
    history = model.fit(x=x_train, y=y_train, 
                        batch_size=32, epochs=in_epochs,
                        shuffle=True, # a good idea is to shuffle input before at each epoch
                        validation_data=(x_valid, y_valid),
                        verbose=1);
    # return weights
    print(model.get_weights())
    
    x_predicted = np.random.uniform(-1, 1, 100)
    y_predicted = model.predict(x_predicted)
    plt.scatter(x_predicted, y_predicted,color='r')
    plt.plot(x_valid, y_target)
    plt.grid(True)
    plt.show()
    return model
    
```

### Exercise 11.3

Da ultimo è stato esteso il modello per eseguire il fit di una funzione bidimensionale; in particolare si è utilizzata la funzione 
$$ f(x,y) = \sin\left(x^2+y^2\right) $$
nell'intervallo $x,y \in [-3/2,3/2]$.  
Rispetto al caso precedente, la rete è stata modificata per accogliere come input un vettore bidimensionale, per il resto le principali caratteristiche della funzione sono rimaste analoghe. La migliore struttura della rete individuata consta di 5 layer con rispettivamente  10, 12, 10, 6 e 2 neuroni ciascuno, tutti con funzione di attivazione tangente iperbolica. L'ottimizzatore è lo *Stochastic Gradient Descent* (sgd) e la funzione costo *Mean Square Error* (mse). Sono state effettuate prove con diversi parametri dell'ottimizzatore e con diverse funzioni costo, ma con esiti poco differenti. L'elemento che realmente si è rivelato imprescindibile è stato l'utilizzo della tangente iperbolica come funzione di attivazione. Senza questa caratteristica la rete non ha mai mostrato una buona capacità di apprendimento.

```python
def do_sin (n_train, n_valid, in_sigma, in_epochs):
    a=4
    b=3
    c=2
    d=3
    # generate training inputs
    np.random.seed(0)
    x_train = 3*np.random.random_sample((n_train, 2))-1.5
    x_valid = 3*np.random.random_sample((n_valid, 2))-1.5
    x_valid.sort()
    print(x_train.shape)
#    print(x_train)
    z_target = np.sin(x_valid[:,0]*x_valid[:,0]+x_valid[:,1]*x_valid[:,1]) # ideal (target) linear function
    
    sigma = in_sigma # noise standard deviation, for the moment it is absent
    z_train = np.random.normal( np.sin(x_train[:,0]*x_train[:,0]+x_train[:,1]*x_train[:,1]), sigma) # actual measures from which we want to guess regression parameters
    z_valid = np.random.normal( np.sin(x_valid[:,0]*x_valid[:,0]+x_valid[:,1]*x_valid[:,1]), sigma)
    
                
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.scatter(x_valid[:,0],x_valid[:,1], z_target, cmap=cm.coolwarm, marker='.')
    plt.grid(True); plt.show()
    
    model = tf.keras.Sequential()
#    model.add(Flatten(input_shape=x.shape[1:]))
    model.add(Dense(10, input_shape=(2,), activation='tanh'))
    model.add(Dense(12, activation='tanh'))
    model.add(Dense(10, activation='tanh'))
    model.add(Dense(6, activation='tanh'))
    model.add(Dense(2, activation='tanh'))
    model.add(Dense(1))
    
    # compile the model choosing optimizer, loss and metrics objects
#    sgd = keras.optimizers.SGD(lr=0.5, momentum=0.9, decay=0.01)
    model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
    # fit the model using training dataset
    # over 10 epochs of 32 batch size each
    # report training progress against validation data
    history = model.fit(x=x_train[:,:], y=z_train,
                        batch_size=32, epochs=in_epochs,
                        shuffle=True, # a good idea is to shuffle input before at each epoch
                        validation_data=(x_valid, z_valid),
                        verbose=1);
    # return weights
    print(model.get_weights())
    
    x_predicted = 3*np.random.random_sample((100, 2))-1.5
    z_predicted = model.predict(x_predicted[:,:])
    
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.scatter(x_predicted[:,0], x_predicted[:,1], z_predicted, color='r', marker='.')
    ax.scatter(x_valid[:,0], x_valid[:,1], z_target)
    plt.grid(True); plt.show()

    return model
    
```
